```python
# Student Management System
# Concepts Used: Lists, Dictionaries, Functions, Classes

class Student:
    def __init__(self, roll_no, name, marks):
        self.roll_no = roll_no
        self.name = name
        self.marks = marks  # Dictionary: {subject: marks}

    def calculate_total(self):
        """Calculate total marks."""
        return sum(self.marks.values())

    def calculate_average(self):
        """Calculate average marks."""
        if len(self.marks) == 0:
            return 0
        return self.calculate_total() / len(self.marks)

    def display_details(self):
        """Display student details."""
        print("\n--- Student Details ---")
        print(f"Roll No: {self.roll_no}")
        print(f"Name: {self.name}")
        print("Marks:")
        for subject, mark in self.marks.items():
            print(f"  {subject}: {mark}")
        print(f"Total Marks: {self.calculate_total()}")
        print(f"Average Marks: {self.calculate_average():.2f}")
        print("------------------------")


# List to store multiple Student objects
students = []


def add_student():
    """Add a new student record."""
    roll_no = input("Enter Roll Number: ")
    name = input("Enter Name: ")

    marks = {}
    num_subjects = int(input("Enter number of subjects: "))
    for i in range(num_subjects):
        subject = input(f"Enter subject {i+1} name: ")
        mark = float(input(f"Enter marks for {subject}: "))
        marks[subject] = mark

    new_student = Student(roll_no, name, marks)
    students.append(new_student)
    print("\n Student record added successfully!")


def display_all_students():
    """Display all student records."""
    if not students:
        print("\n No student records found.")
    else:
        for student in students:
            student.display_details()


def search_student():
    """Search student by roll number."""
    roll = input("Enter Roll Number to search: ")
    for student in students:
        if student.roll_no == roll:
            student.display_details()
            return
    print("\n Student not found!")


# Menu-driven interface
def main():
    while True:
        print("\n===== Student Management System =====")
        print("1. Add Student")
        print("2. Display All Students")
        print("3. Search Student by Roll No")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            add_student()
        elif choice == '2':
            display_all_students()
        elif choice == '3':
            search_student()
        elif choice == '4':
            print("\n Exiting the program. Goodbye!")
            break
        else:
            print("\n Invalid choice! Please try again.")


# Run the program
if __name__ == "__main__":
    main()
```

```python
import numpy as np

# Step 1: Store the sales data using a NumPy array
sales = np.array([85000, 120000, 95000, 130000, 70000, 110000, 90000, 60000, 150000, 100000])

def analyze_sales(sales):
    avg_sales = np.mean(sales)
    print(f"\nAverage Sales: ₹{avg_sales:.2f}\n")

    performance = []  # To store performance category of each salesperson

    for s in sales:
        if s > 1.2 * avg_sales:
            performance.append("Excellent")
        elif s >= 0.8 * avg_sales:
            performance.append("Good")
        else:
            performance.append("Needs Improvement")

    return np.array(performance)

# Step 3: Call the function
performance_array = analyze_sales(sales)

# Step 4: Display results
print("Original Sales Data (₹):")
print(sales)

print("\nPerformance Category for Each Representative:")
print(performance_array)
```

```python
import numpy as np
import pandas as pd

marks = np.random.randint(30, 101, size=10)  # Random integers between 30 and 100

df = pd.DataFrame({
    "Student_ID": np.arange(1, 11),  # IDs from 1 to 10
    "Marks": marks
})

def analyze_performance(df):
    grades = []  

    for mark in df["Marks"]:
        if mark >= 85:
            grades.append("Excellent")
        elif mark >= 70:
            grades.append("Good")
        elif mark >= 50:
            grades.append("Average")
        else:
            grades.append("Fail")

    df["Grade"] = grades

    avg_marks = df["Marks"].mean()
    max_marks = df["Marks"].max()
    min_marks = df["Marks"].min()

    print("\n--- Statistical Analysis ---")
    print(f"Average Marks: {avg_marks:.2f}")
    print(f"Highest Marks: {max_marks}")
    print(f"Lowest Marks : {min_marks}")

    return df

updated_df = analyze_performance(df)

print("\n--- Final Student Performance Data ---")
print(updated_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]

# Generate random monthly sales data for 3 products (in ₹)
np.random.seed(42)  # For reproducible results
product_A = np.random.randint(40000, 100000, size=12)
product_B = np.random.randint(30000, 90000, size=12)
product_C = np.random.randint(20000, 80000, size=12)

# Create DataFrame
df = pd.DataFrame({
    "Month": months,
    "Product_A_Sales": product_A,
    "Product_B_Sales": product_B,
    "Product_C_Sales": product_C
})

df["Total_Sales"] = df["Product_A_Sales"] + df["Product_B_Sales"] + df["Product_C_Sales"]

max_sales_month = df.loc[df["Total_Sales"].idxmax(), "Month"]

avg_A = df["Product_A_Sales"].mean()
avg_B = df["Product_B_Sales"].mean()
avg_C = df["Product_C_Sales"].mean()

print("=== Sales Data ===")
print(df)
print("\nMonth with Highest Total Sales:", max_sales_month)
print("\nAverage Sales per Product:")
print(f"Product A: ₹{avg_A:.2f}")
print(f"Product B: ₹{avg_B:.2f}")
print(f"Product C: ₹{avg_C:.2f}")


# (a) Line chart – monthly sales trends
plt.figure(figsize=(10,6))
plt.plot(df["Month"], df["Product_A_Sales"], marker='o', label="Product A")
plt.plot(df["Month"], df["Product_B_Sales"], marker='o', label="Product B")
plt.plot(df["Month"], df["Product_C_Sales"], marker='o', label="Product C")
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales (₹)")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# (b) Bar chart – total monthly sales
plt.figure(figsize=(10,6))
plt.bar(df["Month"], df["Total_Sales"], color='skyblue')
plt.title("Total Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales (₹)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# (c) Pie chart – annual sales contribution per product
total_A = df["Product_A_Sales"].sum()
total_B = df["Product_B_Sales"].sum()
total_C = df["Product_C_Sales"].sum()

plt.figure(figsize=(6,6))
plt.pie([total_A, total_B, total_C],
        labels=["Product A", "Product B", "Product C"],
        autopct='%1.1f%%', startangle=90, colors=["#66b3ff","#99ff99","#ff9999"])
plt.title("Annual Sales Contribution per Product")
plt.show()
```
